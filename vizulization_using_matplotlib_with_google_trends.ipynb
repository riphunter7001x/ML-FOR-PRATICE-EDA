```python
# get data for ouur analysis
# importing data
from pytrends.request import TrendReq
# create pytrend object and request from google trends
# setting language as english
pytrends = TrendReq(hl="en-us") 
```

```python
# extract data about keywords
# wea re getting data of search frequency for popular languages

keywords = ["python","c++","java","R"]
pytrends.build_payload(keywords,timeframe ="today 5-y")

data = pytrends.interest_over_time()
data
```

```python
data.head()
```

```python
data.tail()
```

```python
type(data)
```

```python
# now plot the data 
import matplotlib.pyplot as plt
plt.plot(data)
# add titles
plt.title("programing languages searches on google")
plt.xlabel("years")
plt.ylabel("weekly search frequency")
# add legend
plt.legend(keywords)
plt.show()
```



*   we got to that past years c++ searches  more on google.
*   java is least searched language

*   r is second highest
